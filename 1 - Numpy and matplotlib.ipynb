There is a lot you can do with 2D arrays, and it will be essential to master this aspect of numpy for handling images, matrices, but also multi dimensional data in general, such as intensity along time, at different z positions.

To begin with, you can create matrices with **zeros**, **ones**, or **diag**.

```python

A = np.zeros(( 4,5 )) #create a 4*5 matrix consisting of zeros. Note that we use matrix convention i.j. So we have 4 lines and 5 rows.
print( A )
print( A.shape ) #Dimensional shape of the object.

A_noised = A + np.random.normal( 1,0.5, A.shape )  #add a gaussian (normal law) noise to A of mean 1 and std 0.5.

print('\n')
print('Adding gaussian noise:')
print( A_noised )

print('\n')
print('Statistics of A :')
print( 'mean :', np.mean(A_noised) )
print( 'std : ', np.std(A_noised) )
```

Here we performed mean and std operation on the whole **A** matrice, however you can specify operations along a given axis.
This is very useful for example to look at z intensity profiles.

```python
Asum = np.sum(A_noised, axis=1) #sum for each line (sums along columns)

print(Asum)

print('\n')
print('Statistics of A:')
print( 'mean along  :', np.mean(A_noised , axis=0) )
print( 'std : ', np.std(A_noised , axis=-1) )
```

You can look back at the matrix A_noised and see how the sum on axis=1 was performed.

### Linear algebra

Numpy's ability to use matrix multiplication opens the way to general linear algebra. For example numpy can invert a matrix with np.linalg.inv()

Thus, we can solve classical problems in the form :  <font size="3">$$ A\cdot x = b  $$
    
Let's see an example :
We will create a profile b which is the sum of three known profiles : 
    
<font size="3">$$ b = 1.5 \cdot a_1 + 2 \cdot a_2 - a_3  $$

The solution of such system is :

<font size="3">$$ (A^{T}A )^{-1} A^{T} \cdot b $$
    


By knowing b and f1, f2, f3, we can estimate each coefficients with the least square method.

```python
#Example of Least Square fit :

arr = np.linspace(0,20, 100) # array on [0,20] with 100 elements.

#we will create a matrix b which is a linear combination of three profiles f1,f2 and f3 : 
a1 = np.sin( arr/5*np.pi)
a2 = 1 + 3*np.exp( -(arr-10)**2/10 )
a3 = np.ones ( arr.shape )

plt.figure('Least Square example')
plt.plot( arr, a1 ,label='a1')
plt.plot( arr, a2 ,label='a2')
plt.plot( arr, a3 ,label='a3')


A = np.vstack( ( f1, f2, f3 ) ) #stack vertically. You could use hstack to stack horizonally.
A = A.T #transposes A.

x = np.array( [1.5, 2, -1] ) #coefficients for linear combination
b = A @ x  # b = f1*x[0] + f2*x[1] + f3*x[2].   @ is matrix multiplication.


plt.plot( arr, b,label=' Linear Mix' )
plt.legend()

xsol = np.linalg.inv( A.T @ A ) @ A.T @ b # @ Least square solve.
print( 'Found coefficients via LSQ : ', xsol )

#in the case where there is for example poisson noise:
b_noised = b + np.random.normal( 0, 0.2 , b.shape )

plt.figure('Least Square example')  # points matplotlib back to the figure with this specific title.
plt.plot( arr, b_noised,label=' Linear Mix + noise' )
plt.legend()
plt.tight_layout() #set a tight image, avoids having too much empty white spaces.

xsol_noised = np.linalg.inv( A.T @ A ) @ A.T @ b_noised # @ 

print( 'Found coefficients via LSQ, with noise : ', xsol_noised )
```

We see that indeed we can get back to the coefficient with high accuracy, even when the signal is impacted by noise !

##### Fourier filtering

As a second example, we can perform Fourier operations on this image with the numpy sub module **fft** \
Fourier transform performs a bijection from a spatial image, to a complex space. The low variying signal is encoded in the center of the complex Fourier space, while high frequencies (linked to high resolution in the image), corresponds to the periphery.


```python
#First we open the Python image and plot its Fourier transform.

import numpy as np
import matplotlib.pyplot as plt
import tifffile as tf

file = "Python_image.tif"
raw_img = tf.imread( file )
gray_img = np.sum( raw_img, axis=-1 )

f_img = np.fft.fftshift( np.fft.fft2( gray_img ) ) #2D fast fourier transform
#by default the center is not at the center of image, fftshift corrects this.
    

plt.figure()
plt.subplot(1,2,1); plt.title('Amplitude');
plt.imshow( np.log( 1 + np.abs(f_img)) );  plt.xticks([]); plt.yticks([]) #Note : the use of ';' allows to write lines consecutively.
plt.subplot(1,2,2); plt.title('Phase')
plt.imshow( np.angle(f_img) ); plt.xticks([]); plt.yticks([])
print('Done')
    
```

As you can see the Fourier transform of our image has high intensity in the center, meaning that most signal corresponds to low varying values. We can build a function that will only keep a specific range of frequencies in the Fourier domain.

```python

#Let's assume we have a slightly noisy image : 
gray_img = np.sum( raw_img, axis=-1 )
gray_img = gray_img + np.random.uniform( 0, 800,  gray_img.shape ) #note: image is now of float type.
print( gray_img.dtype )

#Let's create a function to perform Fourier filtering : we will select specific frequencies in the Fourier image.
#Basically we create a filtering mask whose borders correspond to a specific frequency.
def filter_fourier( fimg, low_freq=0, high_freq=2 , doplot=True, cmap='gray'):
    """ Apply a Fourier filter circle or donut shaped on a complex image fimg. We keep frequencies between low_freq and high_freq.
    Given frequency are in % (0-1) of image radius. 1 will take all image.
    Returns the real (fourier transformed) filtered image """
    fimg = np.fft.fftshift( fimg )
    
    #First we create a 'radius' image
    #here x y are 2D images corresponding to the x position. (you can plt.imshow them if need be )
    x,y = np.meshgrid( np.linspace(-1,1,fimg.shape[1])  ,  np.linspace(-1,1,fimg.shape[0]) )
    radius = np.hypot( x, y )
    filt = (radius>low_freq) * (radius<high_freq)
    filtered_img = filt * fimg
    real_img =  np.fft.ifft2( np.fft.fftshift( filtered_img ) ) #ifft2 = inverse fourier transform.
    real_img = np.real( real_img )
    if doplot:
        plt.subplot(1,3,1); plt.title('Filter Mask')
        plt.imshow(filt, cmap=cmap); plt.xticks([]); plt.yticks([])
        plt.subplot(1,3,2); plt.title('Filtered Fourier Image')
        plt.imshow( np.log(1+np.abs(filtered_img) ) , cmap=cmap) ; plt.xticks([]); plt.yticks([])
        plt.subplot(1,3,3); plt.title('Filtered Real Image')
        plt.imshow( real_img , cmap=cmap) ; plt.xticks([]); plt.yticks([])
        #note: we remove image ticks as they serve no purpose here.
    return real_img
    
    
plt.figure( figsize=(12,8) )
img_filt = filter_fourier( np.fft.fft2(gray_img), low_freq=0 , high_freq=0.2, cmap='gray')

plt.figure( figsize=(12,8) )
plt.subplot(1,2,1); plt.title('Raw image + noise')
plt.imshow( gray_img , cmap='gray')
plt.subplot(1,2,2); plt.title('Filtered image')
plt.imshow( img_filt , cmap='gray')

```

If the image was free of high spatially-varying noise (gaussian noise), keeping only the low frequencies would diminish the contrast on the image and lose a bit of resolution, but smooth the image.
If you want to check this you can apply the above script on an un-noised image and see the result !

In our case here the gaussian noise is a fast spatially-varying noise. It is encoded in the highest frequency of the Fourier space, and is thus efficiently removed by our filter.
So as you can see, Fourier filtering can be used on 1D or 2D data to smooth out the signal, and get rid of specific noises.

We can appreciate the effect of filtering either the low or high frequencies of the image : 

```python

#filtering out highest frequencies : image is smoothed. We lose in resolution but remove fast spatial noise.
img0 = filter_fourier( np.fft.fft2(gray_img), low_freq=0 , high_freq=0.2, doplot=False )

#filtering out high frequencies : image gets blurry
img1 = filter_fourier( np.fft.fft2(gray_img), low_freq=0 , high_freq=0.035, doplot=False )

#filtering low frequencies : image is sharpened, 
img2 = filter_fourier( np.fft.fft2(gray_img), low_freq=0.003 , high_freq=0.5 , doplot=False )

#filtering too many low frequencies : image gets artefacts. mostly contours are kept
img3 = filter_fourier( np.fft.fft2(gray_img), low_freq=0.015 , high_freq=0.2 , doplot=False )

plt.figure( figsize=(12,4))

plt.subplot(2,2,1);
plt.title('Low pass filter')
plt.imshow( img0 );plt.xticks([]); plt.yticks([])
plt.subplot(2,2,2);
plt.title('Very low pass filter')
plt.imshow( img1 );plt.xticks([]); plt.yticks([])
plt.subplot(2,2,3);
plt.title('Band pass filter')
plt.imshow( img2 );plt.xticks([]); plt.yticks([])
plt.subplot(2,2,4);
plt.title('High pass filter')
plt.imshow( img3 );plt.xticks([]); plt.yticks([])
plt.subplots_adjust( wspace=0.1) #minimize spacing between images.
plt.tight_layout()
```

As you can see, Fourier transform is a useful mathematical tool, but you don't have to understand it fundamentally to make it work !

### Image transformations

For image manipulation, such as doing translations and registration, the scipy library can be used.
Here numpy will only serve as a library to handle images and matrices.

Scipy should already be installed with Anaconda.

Below is an example code implementing affine matrix, and using scipy to apply them to our Python image.

```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile as tf

import scipy
import scipy.ndimage #Some sub modules are not automatically loaded with main module. We import them explicitely.

#Creating Affine transformation matrices : 
def affine_scale(x=1,y=1):
    """ return affine matrix corresponding to a x y scaling """
    return np.array( [ [ x, 0,  0], [0, y,0  ] , [0, 0, 1 ] ] )

def affine_shift(x=0,y=0):
    """ return affine matrix corresponding to a x y shift """
    return np.array( [ [ 1, 0,  x], [0, 1,y  ] , [0, 0, 1 ] ] )

def affine_rotate( theta=0, is_degree=False):
    """ return affine matrix corresponding to a theta (rad) rotation """
    if is_degree:
        theta = theta * np.pi/180
    c,s = np.cos(theta) , np.sin(theta)
    return np.array( [ [ c, s,  0], [-s, c, 0  ] , [0, 0, 1 ] ] )

def affine_change_base( M , x0, y0):
    """ return affine matrix M in a translated base of origin (x0,y0) 
    Applied to an image this would rotate it around its top left corner (0,0) """
    P = affine_shift( x0, y0)
    return P @ M @ np.linalg.inv( P)


#Using it on an image :

file = "Python_image.tif"
raw_img = tf.imread( file )
gray_img = np.sum( raw_img, axis=-1 )

#translation
xt,yt = 300, -500
#scale
xs,ys = 1, 0.5
#rotation
thetar = 30 * np.pi/180
xm,ym = [ int(s/2) for s in gray_img.shape ]

#Affine Matrices
Mt = affine_shift(xt,yt)
Ms = affine_change_base( affine_scale(xs,ys) , xm, ym ) #scale around center
Mr = affine_change_base( affine_rotate( thetar ), xm, ym ) #rotation around center

#Images
img_translated = scipy.ndimage.affine_transform( gray_img, Mt )
img_scaled = scipy.ndimage.affine_transform( gray_img, Ms )
img_rotated = scipy.ndimage.affine_transform( gray_img, Mr )

#Plot
plt.figure()
plt.suptitle('Affine transformations with scipy')
plt.subplot(2,2,1)
plt.title('original')
plt.imshow( gray_img , cmap='gray' )
plt.subplot(2,2,2)
plt.title('translated')
plt.imshow( img_translated , cmap='gray' )
plt.subplot(2,2,3)
plt.title('scaled')
plt.imshow( img_scaled , cmap='gray')
plt.subplot(2,2,4)
plt.title('rotated')
plt.imshow( img_rotated , cmap='gray' )
plt.tight_layout()
```

In the future you may try to correct for shift between two images, this can be done via the skimage library (scikit-image). It possesses a function phase_cross_correlation() that returns the shift between two similar images.
